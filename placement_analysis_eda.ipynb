```python
plt.figure()
sns.scatterplot(x='study_hours_per_day', y='cgpa', data=df)
plt.title("Study Hours vs CGPA")
plt.xlabel("Study Hours per Day")
plt.ylabel("CGPA")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='branch', y='cgpa', data=df)
plt.title("CGPA Distribution by Branch")
plt.xlabel("Branch")
plt.ylabel("CGPA")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='projects_completed', y='aptitude_skill_rating', data=df)
plt.title("Projects vs Aptitude Skill Rating")
plt.xlabel("Projects Completed")
plt.ylabel("Aptitude Skill Rating")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
sns.scatterplot(x='cgpa', y='aptitude_skill_rating', hue='branch', data=df)
sns.scatterplot(x='sleep_hours', y='stress_level', data=df)
```

```python
sns.countplot(x='stress_level', data=df)
plt.title("Stress Level Distribution")
plt.show()
```

```python
sns.countplot(x='branch', data=df)
plt.title("Students by Branch")
plt.show()
```
